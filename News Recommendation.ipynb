## Lemmatization and text cleaning

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)   

stem('ordered')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize(text):
    y=[]
    for i in text.split():
        y.append(lemmatizer.lemmatize(i,  pos='v'))
    return " ".join(y)

print(lemmatize('flying'))
```

```python
df['short_description'].iloc[1]
```

```python
df['short_description']= df['short_description'].apply(lemmatize)
df['category']= df['category'].apply(lemmatize)
df['headline']= df['headline'].apply(lemmatize)
```

```python
df['short_description'].iloc[1]
```

```python
def to_lower(text):
    words = text.split(" ")  
    words = [word.lower() for word in words]  
    return " ".join(words)  

print(to_lower("Hello this is News tODAY"))  
```

```python
from nltk.tokenize import RegexpTokenizer

def remove_punctuation_nltk(text):
    tokenizer = RegexpTokenizer(r'[\w\s]+')
    tokens = tokenizer.tokenize(text)
    return ' '.join(tokens)
text = "Hello, world! How's it going? Let's remove everything: @! . and ,."
cleaned_text = remove_punctuation_nltk(text)
print(cleaned_text)
```

## Tf- IDF vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer(stop_words='english', max_features=7000)
vectors = vectorizer.fit_transform(df['tags'])
```

```python
vectors.shape
```

```python
user_input = input("Enter a search term: ").lower()
```

```python
def fetch_headline(index):
    headline= data[index]['headline']
    return headline
```

```python
def fetch_links(index):
    link=data[index]['link']
    return link
```

```python
def fetch_category(index):
    category=data[index]['category']
    return category
```

```python
def recommend(user_input):
    user_input = lemmatize(user_input)
    user_input= remove_punctuation_nltk(user_input)
    user_input= to_lower(user_input)
    user_vector = vectorizer.transform([user_input])
   

    similarity_scores = cosine_similarity(user_vector, vectors)
    top_indices = similarity_scores.argsort()[0][-10:][::-1]  # Top 10 results
    
    for index in top_indices:
        print(fetch_category(index), fetch_headline(index), fetch_links(index))
    
```

```python
recommend(user_input)
```

```python
import joblib

joblib.dump(vectorizer,'vectorizer.pkl')
joblib.dump(vectors, 'vectors.pkl')
```

## SBERT MODEL

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2') 
```

```python
# Encode the merged text column into embeddings (one vector per article)
embeddings = model.encode(df['tags'].tolist(), show_progress_bar=True)

# Check the shape of embeddings (it should be a matrix with 768 dimensions for each article, if using 'all-MiniLM-L6-v2')
print(embeddings.shape)
```

```python
def recommend_sbert(user_input):

    user_input= to_lower(user_input)

    user_vector = model.encode([user_input])
    print(user_vector.shape)

    similarity_scores = cosine_similarity(user_vector, embeddings).flatten()
    top_indices = similarity_scores.argsort()[-10:][::-1]

    for index in top_indices:
        print(fetch_category(index), fetch_headline(index), fetch_links(index))
    
```

```python
recommend_sbert(user_input)
```

```python
model.save('sbert_model')
joblib.dump(embeddings, 'embeddings.pkl')
```

## Evaluation 

```python
with open('feedback.json', 'r') as f:
    data= json.load(f)
    
data
```

```python
def map_user_rating(rating):
    if 7 <= rating <= 10:
        return "Good"
    elif 4 <= rating <= 6:
        return "Moderate"
    elif 1 <= rating <= 3:
        return "Bad"
```

```python
predicted_category=[]
user_ratings=[]

for item in data:
    predicted_category.append(item['avg_classification'])
    user_ratings.append(map_user_rating(item['user_rating']))
    
print(predicted_category)
print(user_ratings)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(user_ratings, predicted_category, labels=["Good", "Moderate", "Bad"])
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Good", "Moderate", "Bad"])

fig, ax = plt.subplots(figsize=(8, 6))
cm_display.plot(ax=ax, cmap="Reds", colorbar=True)
ax.set_xticks(np.arange(-0.5, cm.shape[1], 1), minor=True)
ax.set_yticks(np.arange(-0.5, cm.shape[0], 1), minor=True)
ax.grid(which="minor", color="black", linestyle="-", linewidth=1)
ax.tick_params(which="minor", size=0)

plt.title("Confusion Matrix: User Ratings vs Predicted Classification")
plt.show()
```
